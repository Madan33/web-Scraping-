```python
import os 
import requests
import pandas as pd
import csv
from bs4 import BeautifulSoup
```

```python
url = "https://www.techfetch.com/job-description/python-developer-santa-clara-ca-j3580065"
```

```python
respond = requests.get(url)
```

```python
print(respond)
```

```python
respond.status_code
```

```python
soup = BeautifulSoup(respond.text, "html.parser")
print(soup)
```

```python
job_role= soup.find("div" , class_ = "col-lg-8 col-md-8 col-sm-7").get_text(strip = True)
print(job_role)
```

```python
date = job_role [49:]
print(date)
```

```python
Work_Authorization = soup.find("ul" , class_ = "list-icons").get_text(strip = True)
print(Work_Authorization)
```

```python
Experience= soup.find("div" , class_ = "col-lg-7 col-md-7 col-sm-7 pad0").get_text(strip = True)
print(Experience)
```

```python
s1 = soup.find("div" , class_ = "col-lg-5 col-md-5 col-sm-5 col-lg-pull-1 col-md-pull-1 col-sm-pull-1 col-lg-push-0 col-md-push-0 col-sm-push-1 leftshadow").get_text(strip = True)
print(s1)
```

```python
salary = s1[33:56]
print(salary)
```

```python
Jd= soup.find("span" , class_ = "JobDescCKEditor").get_text(strip = True)
print(Jd)
```

```python
Work_mode = Jd[: 37]
print(Work_mode)
```

```python
Required_skill = soup.find("div", class_ = "col-lg-8 col-md-8 col-sm-8 col-lg-pull-0 col-md-pull-0 col-sm-pull-0").get_text(strip = True)
print(Required_skill)
```
